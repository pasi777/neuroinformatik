```python
# Aufgabe 2.1
import numpy as np
import matplotlib.pyplot as plt

iterations = 10

eps = 10**(-8)
eta = 0.5
beta = 0.9

''' Initialisierung f√ºr Aufgabe 3b)
eps = 0
eta = 0.5
beta = 0
'''

def error(w):
    return -2*(w+ 5)**2 + (w - 6)**3

def error_grad(w):
    return -4*(w + 5) + 3*(w - 6)**2 

def adaptive(eta, beta, iterations):
    w = np.zeros(iterations+ 1)
    w[0] = 4
    s = 0
    for t in range(1,iterations + 1):
        s = beta * s + (1-beta) * (error_grad(w[t-1]))**2
        w[t] = w[t-1] - eta * error_grad(w[t-1]) / (np.sqrt(s + eps))
    return w
```

```python
# Visualisierung

w = adaptive(eta,beta,iterations)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel("w")
ax.set_ylabel("Error")
ax.grid(True)

x = np.linspace(0,iterations + 4,100)
ax.plot(x, error(x), label = "Error function")

ax.scatter(w , error(w) , label = "Updates" , color = "r")

ax.legend();
```
