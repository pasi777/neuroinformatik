```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# initialisierung der Variablen

bob = np.array([1/2,1/4,1/8,1/8])
Entropie_bob = - np.sum(np.dot(bob,np.log2(bob)))
tValues = np.arange(0.01,1,0.01)

# Funktion zum berechnen der Kullback-Leibler-Differenz zwischen Charles und Bob
def D_C_B(t):
    charles = np.array([t*3/4,(1-t)*3/4,1/8,1/8])
    cross_Entropie = - np.sum(np.dot(bob,np.log2(charles)))
    return cross_Entropie - Entropie_bob

# berechne alle KL-Differenzen in Abh√§hingkeit von t
KLdivergences = np.array([D_C_B(t) for t in tValues])

```

```python
# Visualisierung

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(tValues,KLdivergences, label = "KL-Divergence")
ax.scatter(tValues[np.argmin(KLdivergences)],np.amin(KLdivergences), label = "Minimum")
ax.grid(True)
ax.set_xlabel("t")
ax.set_ylabel("D_C(B)")
ax.set_title("Sch√∂ner Plot!")
ax.legend();
```
